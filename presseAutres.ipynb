```python
import pandas as pd
import codecs
import os
from pathlib import Path
from pandas.core.frame import DataFrame
```

```python
corpus_others = pd.read_csv('corpus_others.csv', parse_dates= ['Date'], encoding = "UTF-8")
corpus_others['LenContent'] = corpus_others['Content'].str.len()
corpus_others
```

```python
corpus_en = pd.read_csv('corpus_en.csv', parse_dates= ['Date'], encoding = "UTF-8")
corpus_en['LenContent'] = corpus_en['Content'].str.len()
corpus_en
```

```python
corpus_fr = pd.read_csv('corpus_fr.csv', parse_dates= ['Date'], encoding = "UTF-8")
corpus_fr['LenContent'] = corpus_fr['Content'].str.len()
corpus_fr
```

```python
corpus_all = pd.concat([corpus_fr,corpus_en,corpus_others])
corpus_all
```

```python
import langid
langue = []
for i in corpus_all['Content']:
    lan = list(langid.classify(str(i)))[0]
    langue.append(lan)

langue
```

```python
corpus_all['Language2'] = langue
corpus_all
```
